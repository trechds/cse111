### Case: Score de Crédito dos Clientes

Você foi contratado por um banco para conseguir definir o score de crédito dos clientes. Você precisa analisar todos os clientes do banco e, com base nessa análise, criar um modelo que consiga ler as informações do cliente e dizer automaticamente o score de crédito dele: Ruim, Ok, Bom

```python
# Passo a passo do projeto
# Passo 0 -> Entender o desafio da empresa
# Passo 1 -> Importar a base de dados
# Passo 2 -> Preparar a base de dados para a Inteligência Artificial
# Passo 3 -> Criar um modelo de IA -> Score Crédito: Bom, Ok, Ruim
# Passo 4 -> Escolher o melhor modelo 
# Passo 5 -> Usar a nossa IA para fazer novas previsões
# pandas -> trabalhar com base de dados
# scikit-learn -> inteligencia artificial
# !pip install scikit-learn
```

```python
# Passo 1 -> Importar a base de dados
import pandas as pd

tabela = pd.read_csv('clientes.csv')

display(tabela)
display(tabela.info())
```

```python
# Passo 2 -> Preparar a base de dados para a Inteligência Artificial

# int -> número inteiro
# float -> número com casa decimal
# object -> texto

# Processo de converter texto em número, e depois número no texto
# ***** Label Encoder ***** - profissao, mix_credito, comportamento_pagamento
# Não precisa fazer o Label Enconcer para a coluna que quer prever (nesse caso é o score_credito)
# IA vai treinar com todas as outras colunas para prever a coluna score_credito
# profissao -> 1
# engenheiro -> 2
# cientista -> 3
# advogado -> 4
# artista -> 5
# mecanico -> 6

# importar o label encoder
from sklearn.preprocessing import LabelEncoder

# criar o label enconder
codificador = LabelEncoder()

# aplicar o label encoder na coluna
tabela['profissao'] = codificador.fit_transform(tabela['profissao'])
# tabela['profissao'] = LabelEncoder().fit_transform(tabela['profissao'])

# mix_credito
tabela['mix_credito'] = codificador.fit_transform(tabela['mix_credito'])

# comportamento_pagamento
tabela['comportamento_pagamento'] = codificador.fit_transform(tabela['comportamento_pagamento'])


display(tabela.info())
```

```python
# Passo 3 -> Criar um modelo de IA -> Score Crédito: Bom, Ok, Ruim

# 2 diviões

# Quem eu quero prever, e quem eu quero usar para fazer a previsão

y = tabela['score_credito']
x = tabela.drop(columns=['score_credito', 'id_cliente'])

# treino e teste
from sklearn.model_selection import train_test_split

x_treino, x_teste, y_treino, y_teste = train_test_split(x, y, test_size=0.3)

# overfitting -> quando treinamos a tabela inteira, ela cai no overfitting, que é decorar o conteúdo, e isso é ruim.

# Divisão dos dados em treino e teste (70% treino / 30% teste)
```

```python
# Agora vamos criar o nosso Modelo de IA
    # Modelo 1: Árvore de Decisão
    # Modelo 2: KNN -> Vizinhos Próximos (Nearest Neighbors)

# 3 passos para criar e treinar uma IA
    # Importa a IA
    # Cria a IA
    # Treina na IA

# Importar a IA
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier


# Cria a IA
modelo_arvoredecisao = RandomForestClassifier()
modelo_knn = KNeighborsClassifier()


# Treina na IA
modelo_arvoredecisao.fit(x_treino, y_treino)
modelo_knn.fit(x_treino, y_treino)
```

```python
# Testar modelos
previsao_arvoredecisao = modelo_arvoredecisao.predict(x_teste)
previsao_knn = modelo_knn.predict(x_teste)

# Accuracy

from sklearn.metrics import accuracy_score

print(accuracy_score(y_teste, previsao_arvoredecisao))
print(accuracy_score(y_teste, previsao_knn))
```

```python
# Passo 5 -> Usar a nossa IA para fazer novas previsões
# O melhor modelo é o de Árvore de Decisão -> RandomForestClassifier()

tabela_nova = pd.read_csv("novos_clientes.csv")
display(tabela_nova)

tabela_nova['profissao'] = codificador.fit_transform(tabela_nova['profissao'])
tabela_nova['mix_credito'] = codificador.fit_transform(tabela_nova['mix_credito'])
tabela_nova['comportamento_pagamento'] = codificador.fit_transform(tabela_nova['comportamento_pagamento'])

previsoes = modelo_arvoredecisao.predict(tabela_nova)
display(previsoes)
```
